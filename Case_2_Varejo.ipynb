```python
cliente = pd.read_excel("cliente_varejo.xlsx")
cliente.head()
```

```python
vendas_correto.head()
```

```python
cliente[["idade", "renda"]].describe()
```

```python
cliente.info()
```

```python
cliente = cliente.astype({"renda":float})
cliente.info()
```

```python
vendas_cliente = vendas_correto.merge(cliente, how = "left", on="cliente_Log")
vendas_cliente.head()
```

```python
agg_idcanal_renda = round(vendas_cliente.groupby("idcanalvenda")["renda"].agg("mean").reset_index(),2)
agg_idcanal_renda
```

```python
agg_idade_bandeira = round(vendas_cliente.groupby("bandeira")["idade"].agg("mean").reset_index(),2)
agg_idade_bandeira
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(10,8))
plt.bar(agg_idade_bandeira["bandeira"], agg_idade_bandeira["idade"])
plt.title("Média de Idade por Bandeira")
plt.xlabel("Bandeira")
plt.ylabel("Média de Idade")
plt.show()
```
